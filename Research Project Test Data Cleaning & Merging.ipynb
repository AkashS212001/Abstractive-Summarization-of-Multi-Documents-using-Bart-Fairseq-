```python
with open('test.src', 'r', encoding='utf') as src_file:
    data = src_file.readlines()
```

```python
df = pd.DataFrame(data, columns=['src_content'])
```

```python
with open('test.tgt', 'r', encoding='utf') as tgt_file:
    datatgt = tgt_file.readlines()
```

```python
df1 = pd.DataFrame(datatgt, columns=['tgt_content'])
```

```python
df1
```

```python
# Combine summaries and sources into a single DataFrame
combined_df = pd.concat([df, df1], axis=1)
```

```python
combined_df
```

```python
Truncated_data = combined_df.head(1000)
```

```python
Truncated_data
```

```python
#Removing Paranthesis and Hyphens
Truncated_data['src_content'] = Truncated_data['src_content'].str.replace(r'[\(\)\-]','', regex=True)
```

```python
Truncated_data
```

```python
Truncated_data['tgt_content'] = Truncated_data['tgt_content'].str.replace(r'[\(\)\–]', '', regex=True)
```

```python
Truncated_data
```

```python
def remove_word(Truncated_data, src_content, word):
    Truncated_data['src_content'] = Truncated_data['src_content'].str.replace(word, '')
    return Truncated_data
word_to_remove = 'NEWLINE_CHAR'
column_name = 'src_content'
Truncated_data = remove_word(Truncated_data,column_name,word_to_remove)
```

```python
Truncated_data
```

```python
import re

def remove_emails_websites(text):
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove website links starting with http:// or https://
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)
    
    return text
```

```python
Truncated_data['src_content'] = Truncated_data['src_content'].apply(remove_emails_websites)
```

```python
Truncated_data
```

```python
Truncated_data['tgt_content'] = Truncated_data['tgt_content'].apply(remove_emails_websites)
```

```python
Truncated_data
```

```python
#Removing special character from dataframe
Truncated_data = Truncated_data.replace('[^\w\s]', '', regex=True) 
```

```python
Truncated_data
```

```python
Truncated_data.to_csv('Test_Truncated_Data.csv', index=False)
```
